```python
def divide(a, b):   # fuction divide의 arguments를 정의
  return a/b  # fuction divide의 값을 a/b로 출력해라!
```

```python
divide(10, 0)   # 역시나 a=10, b=0을 대입했을 때, 0으로 나눌 수 없기 때문에 오류 발생! -> ZeroDivisionError: division by zero
```

```python
try:                                               # exception을 처리 할 수 있는 fuction try를 정의. error가 나왔을 때 except로 오류 값에 대한 임의적 출력을 할 수 있다.
  c = divide(5,0)                           # a=5, b=0 , returun 5/0을 하면 [ ZeroDivisionError: division by zero ] 가 출력되야 한다.
except:                                         # 하지만 except fuction을 정의 하면 error 코드 대신 예외 사항으로 임의 코드를 쓸 수 있다!
  print('Exception is occured!!')   # 오류가 발생하면, 'Exception is occured!!' 를 출력해주세요!
  
# try:
#     실행할 코드
# except:
#     예외가 발생했을 때 처리하는 코드
# https://dojang.io/mod/page/view.php?id=2398
```

```python
try: 
  c = divide(10,0)  # 10을 a으로 나누는 함수..
except ZeroDivisionError:                             # 만약 ZeroDivisionError 에 대한 오류가 나온다면~
  print('두 번째 값은 0이 되면 안됩니다. ')  # 이 말을 출력해주세요
except TypeError:                                         # 만약 TypeError 에 대한 오류가 나온다면~
  print('모든 인자는 숫자여야 합니다.')       # 이 말을 출력해주세요
except:
  print('알 수 없는 오류입니다.')                 # ZeroDivisionError  / TypeError에 대한 오류가 아니라면 이 말을 출력해주세요
finally:
  print('감사합니다.')  # 그리고 마지막에 이 말을 출력해주세요
```
