```python
class Person:   # Person class 선언
  Name = ''   # Name attribute 정의, Default 값은 없음
  Age = 0     # Age attribute 정의, Default 값은 0

  def __init__(self, name, age):    # 생성자 method : __init__을 정의, instance의 안에는 name과 age의 attribute 값을 순서대로 {0},{1}에 지정할 수 있다.
    self.Name = name                # 생성된 instance에 Name이란 속성으로 name값을 넣고,
    self.Age = age                      # Age란 속성으로 age값으로 넣는다.

  def __del__(self):                  # 소멸자 method : __del__을 정의
    print('{0}, 나이{1} 잘 살다 갑니다ㅜㅜ.'.format(self.Name, self.Age))    # 추가로 지정한 instance의 Name 속성을 {0} index에 넣고, Age 속성을 {1} index에 넣어 출력한다.
    

  def PrintName(self):          # PrintName method의 정의
    print(self.Name)              # instance의 Name 속성 출력
    print(self.Age)                 # instance의 Age 속성 출력
```

```python
p = Person('황비홍',40)     # p를 Person의 isntance로 지정, 속성 Name의 값  name으로  '황비홍' 입력 / 속성 Age의 값 age로 40 입력
p.PrintName()                   # instacne p의 PrintName method 실행 -> instance P의 속성인 Name과 Age의 값(name,age)이 출력 된다.
del p                                 # intacne p를 제거 -> 소멸자 method를 정의했기 떄문에, __del__의 print 값이 함께 출력됨  
```

```python
# 정적(Static) 메소드 

class CounterManager:   # ConterManager class 선언
  InsCount = 0                # InsCount라는 속성의 값을 0으로 초기화(Default값 = 0)

  def __init__(self):           # 생성자 method : __init 정의
    CounterManager.InsCount += 1    # instance를 생성할 때 마다, ConterManger class의 속성인 InsCount의 값이 1씩 늘어난다. (초기화 값 0)
    
  def PrintInstanceCount():   #printInstanceCount 라는 metod의 정의, self를 제외했기 때문에, intance로 method 호출이 불가!
    print('Instance Count:', CounterManager.InsCount)   # 속성 InstanceCount의 값을 print로 출력
  
  StaticCount = staticmethod(PrintInstanceCount)    #정적 메소드로 전환을 명시하는 코드 , 이제부터 PrintInstanceCount method는 static method(정적 메소드)가 된다.
```

```python
# 상속의 심화

class Person:   # Person class 선언

  def __init__(self, name, phone):    # instance 생성자 method 정의
    self.Name = name                     # 속성 Name의 입력 값 = name
    self.Phone = phone                   # 속성 Phone의 입력 값 = phone
  
  def PrintInfo(self):    # PrintInfo method 정의
    print('Info(Name: {0}, Phone {1})'.format(self.Name, self.Phone))  # instance의 속성 Name,Phone의 값 name,phone을 {0},{1} index로 출력
  
  def PrintPersonData(self):    # PrintPersonData method 정의
    print('Person(Name: {0}, Phone {1})'.format(self.Name, self.Phone))  # instance의 속성 Name,Phone의 값 name,phone을 {0},{1} index로 출력
  
```

```python
class Student(Person):    # Student class를 Person class의 자녀 class로 선언
  def __init__(self, name, phone, subject, studentID):  # Student class의 생성자 method 선언
    self.Name = name
    self.Phone = phone
    self.Subject = subject
    self.StudentID = studentID
```

```python
p = Person('박보검','010-1111-1111')    # intance p 생성, Name 속성 값 = '박보검', Phone 속성 값 = 010-1111-111 입력
s = Student('민아','010-2222-2222','학생회장','0001')   #   instance s 생성, Name / Phone / Subject / Student ID 속성 값 입력
```

```python
print(p.__dict__)   # instance p의 속성 값을 dictionary type으로 출력
print(s.__dict__)   # dictionary type이기 때문에 {key : value} 형태로 출력된다!
```

```python
s.PrintInfo()   
# instance s는 Person class에 속해 있기 때문에, PrintInfo method 호출이 가능하다!
# 하지만 PrintInfo method의 정의를 속성 Name과 Phone만 출력하기 때문에, name 값과 phone 값만 출력!
```

```python
# 상속관계를 확인 [ issubclass( Subclass:하위클래스 , SuperClass상위클래스 ) ]
issubclass(Student, Person) # Student class는 Person class의 하위 class인가? -> True
```

```python
issubclass(Person, Student) # Person classsms Student class의 하위 class 인가? -> false
```

```python
class Student(Person):
  def __init__(self, name, phone, subject, studentID):
    self.Name = name
    self.Phone = phone
    self.Subject = subject
    self.StudentID = studentID

    Person.__init__(self, name, phone)

  def PrintStudentData(self):
    print('Student(Name : {0}, StudentID : {1})'.format(self.Name, self.StudentID))
  
  def PrintPersonData(self):
    print('{0}님의 전공은 {1}입니다.'.format(self.Name, self.Subject))
```

```python
# 다중 상속

class Tiger:
  def Jump(self):
    print('호랑이 점프')
```

```python
class Lion:
  def Bite(self):
    print('사자의 물기')
```

```python
class Liger(Tiger, Lion):   # Liger class는 Tiger class와 Lion class을 모두 상속 받는다!
  def Play(self):
    print('라이거의 놀기')
```

```python
l = Liger() # instance l 지정
print(l.Jump())
print(l.Bite())
print(l.Play())
# Liger class는 Lion class, Tiger class 모두를 상속 받기 때문에 두 class의 method인 Jump와 Bite를 모두 호출 가능!
# 하지만, method 정의를 하면서 속성 값을 아무것도 입력 받지 않았기 때문에, None 이라는 속성 값이 출력된다..! 
```
