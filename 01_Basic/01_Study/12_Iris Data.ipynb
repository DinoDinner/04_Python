```python
# Iris Flower Data

from sklearn.datasets   import  load_iris
#   sklearn : machine learing module
#   datasets : iris 꽃 data가 들어있는 set

iris = load_iris()
features = iris.data.T  # T는 data를 Tranpose를 해준다.
features[0]     # 0번째 특성을 가지고 온다.

#   Iris의 Feature : 꽃의 품좀을 구별할 수 있는 가장 기본적인 특성
#   sepal width (cm)
#   petal length (cm)
#   petal width (cm)    
#   데이터를 구성할 땐 Feture와 Label로 구성된다.
#   Machine learning 하기 전에 꼭 labeling(그 특성이 ~다) 이 되어있는게 중요하다.

# [ 덩어리마다 ] 각 feature를 가지고(총 4개), 데이터를 가지고 있다.
```

```python

# matplotlib module 불러오기!

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# Iris Feature로 그래프 생성

plt.scatter(features[3], features[1], c=iris.target, alpha=0.1, s=features[3]*100, cmap='viridis')
#   0번과 1번 feature로 좌표가 찍힌다!
#   c=iris.target / 꽃의 종류마다 색상을 다르게 설정.
#   s=features[3]*100 / 3번 feature의 크기를 3으로 지정하고 100만큼 키움

plt.xlabel(iris.feature_names[0])   #   0번 feature의 label 이름 붙이기
plt.ylabel(iris.feature_names[1])   #   1번 feature의 label 이름 붙이기

iris.target_names[0]    # setosa라고 표기됨!

#   sklearn에서는 label을 target이라고 부른다.
#   iris 3종류의 값이 0,1,2로 표기됨

#   데이터가 3블록으로 나눠지지만, 분리가 더 잘 될수록 좋은 데이터다.
#   데이터가 겹치는 부분 = 노이즈

#   alpha 값을 낮추고, 색이 겹치는 곳이 많을 수록 데이터의 중심!
```

```python
# 다른 Feature로 생성하기

from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[2],features[3],c=iris.target,s=features[3]*100,alpha=0.2,cmap='viridis')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
iris.feature_names[3]

# 좀 더 덩어리화가 잘되었고, 데이터로 활용성이 높아졌다.
```

```python
# 오차 시각화 하기 

x = np.linspace(0,10,50)
dy = 0.8    # 오차범위 0.8로 지정
y = np.sin(x + np.random.randn(50))
# 0을 기준으로 +- 인 랜덤 숫자이기 때문에 분산이 커진다.

plt.errorbar(x,y, yerr=dy ,fmt='.k' )   
# 오차 범위에 대해 시각화 해준다.
# yerr : y축(세로)으로 오차범위 에러를 표시한다.
# fmt(format String) : 점으로 표기하고, 검은색으로 하겠다.
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='k', ecolor='lightgrey', elinewidth=3)
#   ecolor : 에러의 범위 색깔을 변경
#   elinewidth : 에러의 범위 선굵기를 변경

plt.title('error chart') 
```
