```python
# 신경망 기초 

import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1차함수 : Linear Fuction

def linear_function(x): # 함수 정의
    a=0.5
    b=2
    
    return a*x+b

print(linear_function(5))   # 값 도출

# 그래프 출력

x = np.arange(-5,5,0.1)
y = linear_function(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear_function')
```

```python
# 2차함수 : Quradratic Fuction

def quadratic_function(x):  # 함수 정의
    a = 1
    b = -1
    c = -2
    
    return a*x**2 + b*x + c
print(quadratic_function(5))    # 값 도출

# 그래프 출력

x = np.arange(-5,5,0.1)
y = quadratic_function(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic_function')
```

```python
# 3차함수 : Cubic Function

def cubic_function(x):  # 함수 정의
  a = 4
  b = 0
  c = -1 
  d = -8

  return a*x**3 + b*x**2 + c*x + d

x = np.arange(-5,5,0.1)
y = cubic_function(x) # 값 도출

# 그래프 출력

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function')
```

```python
# Tensor(다차원 데이터 배열)의 표현
# Tensor는 기본적으로 Numpy 배열을 사용한다.

x = np.array(3)
print(x)    # 배열 출력
print(x.shape)  # 배열의 형태 (행,열)
print(x.ndim)   # 배열의 차원 N
```

```python
# 1D(일차원) Tensor

a = np.array([1,2,3,4])
b = np.array([5,6,7,8])

c = a + b
d = a + x   # 1차원과 0차원 사이의 연산

print(d)
print(d.shape)
print(d.ndim)

# 차원이 달라도 서로 연산이 된다.
# Numpy Broadcating : 모양이 다른 배열들 간의 연산이 어떤 조건을 만족했을 때 가능해지도록 배열을 자동적으로 변환해줌
```

```python
# 1D 연산

x = np.array([1,2,0])
y = np.array([0,2,1])
z = np.dot(x, y)    # dot : 행렬 곱셈 연산, (1*0) + (2*2) + (0*1) = 4

print(z)
print(z.shape)
print(np.ndim(z))
```

```python
# 2D(이차원) Tensor

matrix = np.array([[1,2,3],[4,5,6]])

print(matrix)
print(matrix.shape)
print(np.ndim(matrix))
```

```python
# 2D 연산

a = np.array([[1,2],[3,4]])
b = np.array([[10,10],[10,10]])

print(a*b)  # 2차원 끼리 자리 찾아서 계산이 된다.
print(np.shape(a*b))
```

```python
# 3D(삼차원) Tensor

x = np.array([
            [[5,3,2,1],
             [5,5,3,1],
             [6,1,2,3]],
            [[1,1,1,1],
             [3,4,7,5],
             [1,8,3,4]],
            [[10,9,3,9],
             [5,4,3,2],
             [7,6,3,4]]
            ])

print(x)
print(x.shape)
print(np.ndim(x))
```
