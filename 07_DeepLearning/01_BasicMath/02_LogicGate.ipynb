```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 논리 게이트
# Percept 론 : 물리적인 w값, bias 값만 변화하면 여러 논리 게이트를 만들 수 있음.

# AND
def AND(a,b):
    input = np.array([a,b])
    weights = np.array([0.4,0.4])
    bias = -0.6
    value = np.sum(input * weights) + bias
    
    if value <= 0:
        return 0
    else:
        return 1
    
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

```python
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias = -0.6

y = (-0.4 * x1 - bias) / 0.4

plt.axvline(x=0)
plt.axhline(y=0)

plt.plot(x1, y,'r--')
plt.scatter(0,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(1,1,color='black',marker='^',s=150)

plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
```

```python
# OR

def OR(a,b):
    input = np.array([a,b])
    weights = np.array([0.4,0.4])
    bias = -0.3
    value =np.sum(input * weights) + bias
    
    if value <=0:
        return 0
    else:
        return 1
    
print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))

# 가중치를 조절하면 새로운 결과 값이 나온다. (bias 조절)
```

```python
# NAND

def NAND(a,b):
    input = np.array([a,b])
    weights = np.array([-0.5,0.5])
    bias = 0.7
    value =np.sum(input * weights) + bias
    
    if value <=0:
        return 0
    else:
        return 1
    
print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))

x1 = np.arange(-2, 2, 0.01)
x2 = np.arange(-2, 2, 0.01)
bias = 0.7

y = (0.5 * x1 - bias) / -0.5

plt.axvline(x=0)
plt.axhline(y=0)

plt.plot(x1, y, 'r--')
plt.scatter(0,0,color='black',marker='o',s=150)
plt.scatter(0,1,color='black',marker='o',s=150)
plt.scatter(1,0,color='black',marker='o',s=150)
plt.scatter(1,1,color='orange',marker='^',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
plt.show()

# 가중치를 조절하면 새로운 결과 값이 나온다. (bias, weight 조절)
# 뉴련이 작용하는 것과 비슷함. 사람도 가중치(bias)값에 따라 변화
```

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)

  return y

print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```
