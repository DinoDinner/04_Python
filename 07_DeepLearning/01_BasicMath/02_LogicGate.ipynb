```python
# 논리 게이트
# Percept 론 : 물리적인 w값, bias 값만 변화하면 여러 논리 게이트를 만들 수 있음.

import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# AND Gate

def AND(a,b):
    input = np.array([a,b])
    weights = np.array([0.4,0.4])   # 가중치는 a에 0.4 / b에 0.4
    bias = -0.6 # 편차 : -0.6
    value = np.sum(input * weights) + bias  # 결과 값
    
    if value <= 0:
        return 0
    else:
        return 1
    
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1)) # 결과적으로 (1,1)만 1이 도출되기 때문에 AND Function 성립
```

```python
# OR Gate

# AND Gate에서 Bias값만 수정.
def OR(a,b):
    input = np.array([a,b])
    weights = np.array([0.4,0.4])
    bias = -0.3
    value =np.sum(input * weights) + bias
    
    if value <=0:
        return 0
    else:
        return 1
    
print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))

# 가중치를 조절하면 새로운 결과 값이 나온다. (bias 조절)
# AND Fuction의 경우 bias = -0.6 이었지만, bias=-0.3일 때 OR Function으로 변화함!
```

```python
# NAND Gate

# 기존의 Gate에서 weights와 bias 수정.
def NAND(a,b):
    input = np.array([a,b])
    weights = np.array([-0.5,-0.5])
    bias = 0.7
    value =np.sum(input * weights) + bias
    
    if value <=0:
        return 0
    else:
        return 1
    
print(NAND(0,0))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(1,1))
```

```python
# XOR Gate

def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)

  return y

print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```

```python
# AND Gate Graph

# 함수 정의
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias = -0.6
y = (-0.4 * x1 - bias) / 0.4

# 기준선 만들기
plt.axvline(x=0)
plt.axhline(y=0)

plt.plot(x1, y,'r--')   # 함수 그래프 -> 선을 넘어가면 1
plt.scatter(0,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='orange',marker='s',s=150)
plt.scatter(1,0,color='orange',marker='d',s=150)
plt.scatter(1,1,color='black',marker='^',s=150) # 검은 삼각형(1,1) 만 선을 넘어서 통과.

# 그래프 옵션
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
plt.title('AND GATE')
```
