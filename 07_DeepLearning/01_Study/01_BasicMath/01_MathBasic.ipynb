```python
# 신경망 기초 

import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1차함수
def linear_function(x):
    a=0.5
    b=2
    
    return a*x+b

print(linear_function(5))   # 값 도출

# 그래프
x = np.arange(-5,5,0.1)
y = linear_function(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear_function')
```

```python
# 2차함수
def quadratic_function(x):
    a = 1
    b = -1
    c = -2
    
    return a*x**2 + b*x + c
print(quadratic_function(5))    # 값 도출

# 그래프
x = np.arange(-5,5,0.1)
y = quadratic_function(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic_function')
```

```python
# 3차함수
def cubic_function(x):
  a = 4
  b = 0
  c = -1 
  d = -8

  return a*x**3 + b*x**2 + c*x + d

x = np.arange(-5,5,0.1)
y = cubic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cublic Function')
```

```python
# 1차원 연산

x = np.array([1,2,0])
y = np.array([0,2,1])
z = np.dot(x, y)    # 모든 값을 합해서 하나의 점으로 만듬

print(z)
print(z.shape)
print(np.ndim(z))
```
