```python
# 세트(set)
a = {1,2,3}
b = {3,4,5}
```

```python
print(a, b)
print(type(a))
```

```python
# 합집합, 교집합을 구할 수 있다. 
print(a.union(b))         # 합집합
print(a.intersection(b))  # 교집합
```

```python
print(a-b)                # 차집합
print(a|b)                # 합집합
print(a&b)                # 교집합
```

```python
# 튜플(tuple)
t = (1,2,3)
print(type(t))
```

```python
(a, b) = (1, 2)
print(a, b)
```

```python
a = 10
b = 20
(a, b) = (b, a) 
print(a, b)
```

```python
# 타입의 전환도 무척 간단하다. 
a = set((1,2,3))
print(type(a))
b = list(a)
print(type(b))
c = tuple(b)
print(type(c))
d = set(c)
print(type(d))
```

```python
# in 연산자를 통해서 해당 값이 있는지 확인 가능하다. 
a = set((1,2,3))
print(1 in a)         # in 연산자는 데이터 타입에 상관없이 사용할 수 있다. 
```

```python
# 사전 (Dictionary)

d = dict(a=1, b=2, c=3)
print(d)
print(type(d))
```

```python
color = {'apple':'red', 'banana':'yellow'}
print(color)
```

```python
print(color['apple'])
#print(color[0])         # dict 타입은 인덱스 없습니다. 
#print(color['cherry'])   # 없는 것을 물어도 예외가 발생된다.
```

```python
color['cherry'] = 'red'
print(color)
```

```python
color['apple'] = 'green'
print(color)
```

```python
print(color.keys())
print(color.values())
print(color.items())
```

```python
# for 문을 사용해서 데이터 가져와보기
for k in color.keys():
  print(k)
for tmp in color.values():
  print(tmp)
for i in color.items():
  print(i)
for (k, v) in color.items():
  print(k, v)
```

```python
# Dict에서 불필요한 아이템의 삭제 

del color['cherry']
print(color)
```

```python
color.clear()
print(color)
```

```python
# 데이터 타입은 서로 혼용해서 사용하는 것이 가능하다. 
d = {'age': 40.5, 'job': [1,2,3], 'name': {'Kim':2}, 'Cho': 1}
print(d)
```

```python
# bool 타입
isRight = False
print(type(isRight))
```

```python
print(1<2)
print(1==2)
print(1!=2)
```

```python
# 논리 연산자
print(True and False)
print(True & True)
print(True or False)
print(False | False)
print(not True)
```

```python
# 얕은 복사와 깊은 복사

a = [1,2,3]
b = a
print(b)
```

```python
a[0] = 38
print(a)
print(b)
```

```python
# Call by value
a = [1,2,3]
b = a[:] 
a[0] = 38
print(a, b)
```

```python
# copy module을 사용하는 법
import copy

a = [1,2,3]
b = copy.deepcopy(a)
a[0] = 38
print(a, b)
```
