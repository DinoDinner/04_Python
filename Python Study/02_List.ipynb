```python
# List
# Python type 중 가장 무겁고 기능이 많은게 List type이다.

colors = ['red', 'green', 'gold']
print(colors)
print(type(colors))
```

```python
# List의 메모리 안에 데이터 추가 : append

colors.append('blue')
print(colors)
```

```python
# 자신이 원하는 위치에 데이터를 추가(삽입) 할 수 있음. : insert

colors.insert(1,'black')
print(colors)
```

```python
# 여러개의 데이터를 추가할 수 있음 : extend -> 뒤에 여러개를 붙일 수 있음.

colors.extend(['white','gray'])
print(colors)
```

```python
# 사칙연산 기호로 데이터 추가 a+=[b], 혼란스러울 수 있으니 사용하지 않는게 좋음!

colors += ['red']
print(colors)
```

```python
#  [ ] 연산자를 통해 추가 할때, 괄호가 없으면 문자가 하나씩 들어감.
# 'red' =-> r, e, d 하나씩 추가됨
# [ ] : 꺽쇠 -> 리스트를 나타내는 괄호

colors += 'red'
```

```python
print(colors)
```

```python
# 특정 데이터의 순번을 알고싶을 때  : index
# 특정 데이터의 중복 개수를 알고 싶을 때 : count

print(colors.index('red'))
print(colors.index('red', 1))   # 0번의 red 이후 1번의 red 순번
print(colors.count('red'))
```

```python
# List에서 데이터 추출 : pop -> 뒷 순번 부터 하나씩 제거

colors.pop()
```

```python
colors.pop()
```

```python
colors.pop(1)
```

```python
print(colors)
```

```python
# remove는 해당 값을 삭제 할 수 있습니다. 

colors.remove('gold')
print(colors)
```

```python
# sort는 알파벳 순서대로 정렬한다. 

# 정순
colors.sort()
print(colors)

#역순
colors.reverse()
print(colors)
```

```python
# 타입의 전환도 무척 간단하다. 

a = set((1,2,3))
print(type(a))

b = list(a)
print(type(b))

c = tuple(b)
print(type(c))

d = set(c)
print(type(d))
```

```python
# in 연산자를 통해서 해당 값이 있는지 확인 가능하다. 

a = set((1,2,3))
print(1 in a)         # in 연산자는 데이터 타입에 상관없이 사용할 수 있다. 
```

```python
# 사전 (Dictionary)
# 다른 Type들과 다르게 Key값과 Value 값이 존재함!

d = dict(a=1, b=2, c=3)
print(d)
print(type(d))
```

```python
color = {'apple':'red', 'banana':'yellow'}
print(color)
```

```python
#print(color[0])         # dict 타입은 인덱스 없습니다. 
#print(color['cherry'])   # 없는 것을 물어도 예외가 발생된다.

print(color['apple'])
```

```python
# DIct의 데이터 추가

color['cherry'] = 'red'
print(color)
```

```python
# DIct의 Value 데이터 수정

color['apple'] = 'green'
print(color)
```

```python
# DIct의 데이터 출력 방식 : keys, values, items -> 뒤에 s 붙이기!

print(color.keys())
print(color.values())
print(color.items())
```

```python
# for 문을 사용해서 데이터 가져와보기
# for 문을 사용하면 모든 데이터가 하나씩 추출된다. ( 연속 추출 )
# 'for' a 'in' b 의 형태

for k in color.keys():
  print(k)

for tmp in color.values():
  print(tmp)

for i in color.items():
  print(i)

for (k, v) in color.items():
  print(k, v)
```

```python
# Dict에서 불필요한 아이템의 삭제  : del

del color['cherry']     # DIct List에서 'cherry'를 삭제한다
print(color)
```

```python
color.clear()
print(color)
```

```python
# 논리 연산자
print(True and False)   # 1 * 0 = 0
print(True & True)      # 1*1 = 1
print(True or False)    # 1 + 0 = 1
print(False | False)    # 0 - 0 = 0
print(not True)         # 0이 아니라면 1
```

```python
# 얕은 복사와 깊은 복사
# 얕은 복사 : 데이터의 주소를 복사 ( 바로가기 )
# 깊은 복사 : 데이터의 값을 복사 ( 복사 )

a = [1,2,3]
b = a       # a의 '바로가기'는 b
print(b)    # b=a를 출력
```

```python
# a의 0열의 값을 수정해도, b는 a의 바로가기 이기 때문에 값이 똑같이 수정된다.

a[0] = 38
print(a)
print(b)
```

```python
# Call by value : a의 List를 복사함 -> 깊은 복사

a = [1,2,3]
b = a[:] 

a[0] = 38
print(a, b)
```

```python
# copy module을 사용하는 법

import copy     # copy 라는 모듈을 import해서 불러옴.

a = [1,2,3]
b = copy.deepcopy(a)    # copy module 안의 deepcopy method 사용
a[0] = 38
print(a, b)
```
