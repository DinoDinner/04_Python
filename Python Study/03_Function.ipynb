```python
# Function 함수
# 함수란, 매개변수에 의해 결과값이 도출되야함.

def Times(a, b):    # Times라는 함수를 정의한다.
  result = a * b     # Times 라는 함수의 기능을 설정
  return result      # Times 라는 함수의 결과 도출을 설정
```

```python
# Times 라는 함수의 기능을 실행
# 하지만, argument가 정해지지 않았기 때문에 Times라는 함수가 메모리에 저장된 위치만 출려됨

print(Times)
```

```python
# Parameter 입력
# 입력된 Parameter 10, 20은 Instance(Times)의 Argument가 된다.
print(Times(10, 20))
```

```python
# 메모리 주소를 확인하는 명령
# 추출된 주소 0x7fcf853e5c10 와 140529270414352 은 사실 같은 값이다.
print(id(Times))
```

```python
# Python의 전역 변수를 확인하고 싶을 때 사용하는 명령어 globals
# 전역 변수는 global variable
# 지역 변수는 local variable

globals()
```

```python
# Times라는 Instance를 다시 정의

myTimes = Times
print(myTimes(10,20))
```

```python
# 정의를 했기 때문에, 다시 출력해도 같은 주소가 나온다.

print(id(myTimes))
```

```python
#  Parameter(x,y)와 Argument(a,b)를 다르게 설정해도 작동한다.
# x,y는 변할 수 있는 값 = 변수
# a,b는 값을 정의 한 것, 변할 수 없음 = 인자

# a와 b의 값을 정의
a = 10
b = 20

# x,y는 어떤 값이 들어올지 정해져 있지 않다. 하나의 대명사
def sum1(x, y):
  return x+y
```

```python
print(sum1(a,b))
```

```python
# 변수의 범위 : 전역 변수(global variables)와 지역 변수(local variables)

g = 10
x = 20

def sum2(x, y):   # x,y parameter는 sum2라는 함수 속에서만 정의된 지역 변수
  global g    # g 라는 parameter를 현재 들어가 있는 함수와 상관없는 전역 변수로 정의 
  g = g+1
  return x, y

print(sum2(x, 20))
print(g)
```

```python
# 함수의 인자 (Parameter)

def Times(a=10, b=20):      # 디폴트 값을 지정
  return a * b

print(Times(100, 200))
print(Times(100))
print(Times())
print(Times(b=200))
```

```python
# URL에 접속하는 함수 만들기
# 기본적인 web site의 포트 번호는 80이다.

def connectURL(protocol='https', server='naver.com', port='80'):
  address = protocol + '://' + server + ':' + port
  return address

print(connectURL(port='8080'))    # port number 8080 할 경우 안들어가지는 사이트도 있음 -> 사이트 재량
```

```python
# 가변인자 함수
# parameter에 *을 붙이면, 가변적이게 정의 할 수 있다.
# Insatance를 여러개 설정 할 수 있음. = 가변 인자(Variable Instance)

def test(*args):
  print(type(args))
  print(args)

test(1,2,3,4,5)
test(1,2,3)
```
