```python
f = open('test.txt','w')    # intance f는 text.txt 라는 파일에 작성을 할 수 있는 w(write) 기능을 호출 할 수 있다!
# 실행하면, text.txt 파일이 없을경우 파일 생성해줌!

# instacne = open(file, mode) 
# https://www.w3schools.com/python/ref_func_open.asp
# "r" - Read - Default value. Opens a file for reading, error if the file does not exist
# "a" - Append - Opens a file for appending, creates the file if it does not exist
# "w" - Write - Opens a file for writing, creates the file if it does not exist
# "x" - Create - Creates the specified file, returns an error if the file exist
```

```python
print('Welcome to python', file = f)    # opne('text.txt', 'w') method를 사용해서 'Welcome to python' 이라는 텍스트를 입력할 수 있다!
```

```python
f.close()   # 하지만, text.txt 파일을 열어서 작성만 되었을 뿐, 파일을 저장해야지 문자가 입력된 화면을 볼 수 있다.
# instance.close() -> instacnce로 호출한 fuction을 마무리한다.
```

```python
f = open('test.txt', 'r')   # text.txt 를 열어서 read 할 수 있는 method 호출할 수 있다!
```
