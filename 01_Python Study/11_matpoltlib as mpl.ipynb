```python
# matplotlib module 불러오기!

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Style의 지정
plt.style.use('classic')    # classic 스타일 적용
```

```python
# 임의의 숫자 생성

import numpy as np  # numpy 호출
x = np.linspace(0,10,100)  # 원하는 범위의 숫자를 필요한 개수만큼 생성할 수 있음

# x라는 변수에 저장.
# 0~10 사이의 100개의 숫자를 만들어 주세요 -> 알정하게 100개를 만들어줌

x
```

```python
# 데이터 시각화

plt.plot(x, x)  # x축 2개를 그리기
sinprint(np.sin(x))
```

```python
# sin(x) 그래프 그리기

plt.plot(x, np.sin(x)) # x축 하나, numpy의 sin(x) 축 하나.

# x축은 0~10까지 계속 증가.
# np.sin(x)는 값이 왔다 갔다...
```

```python
# cos (x)그래프 그리기

plt.plot(x, np.cos(x))  # x축, 하나 numpy의 cos(x) 축 하나
```

```python
# 한 그래프에 sin,cos 같이 표기하기!

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
# jupyter notebook은 한 셀마다 프로그램이 작동하기 때문에 한 셀에 만들기
```

```python
# 그래프를 img로 저장하기

fig = plt.figure() # 그림판 method : figure
plt.plot(x, np.sin(x), '-')     # - : 하나의 실선, 기본값 = solid line
plt.plot(x, np.sin(x), '--')    # -- : -는 선의 모양을 다르게 만들 수 있음!
fig.savefig('./99_Fig/my_figure.png')    # 그림판으로 저장! -> 99_Fig에 위치지정.
```

```python
# 한번에 두 개의 그래프를 출력하는 방법

plt.figure() # 그림판 하나 만들기
```

```python
# 그래프의 레이아웃 지정하기

plt.subplot(2,1,1)   # rows, column, panel number -> 그래프의 레이아웃 지정하기
plt.plot(x, np.sin(x))  # 1번 subplot 그래프에 sin(x)

plt.subplot(2,1,2)
plt.plot(x, np.cos(x)) # 2번 subplot 그래프에 cos(x)
```

```python
# 그래프의 color 지정

plt.plot(x, np.sin(x), color='blue')
plt.plot(x, np.sin(x-1), color='g') # 단축어
plt.plot(x, np.sin(x-2), color='0')  # 0이면 검은색, 0이면 흰색
plt.plot(x, np.sin(x-3), color='#FFdd44')   # HEX code
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3))   # CMY를 숫자로 비율 넣기
plt.plot(x, np.sin(x-5), color='chartreuse')    # 색상의 이름을 넣기
```

```python
# 그래프의 linestyle 지정

# 문자로 표시
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

# 기호로 표시
plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
#  짧은 방식으로 그래프 스타일 지정

plt.plot(x, x+0, '-g')        # solid / green 을 한 번에 표기 가능하다!
plt.plot(x, x+1, '--c')       # dashed / cyan
plt.plot(x, x+2, '-.k')       # dashdot / black
plt.plot(x, x+3, ':r')         # dotted / red
```

```python
# 세부조정 하는 법

# 그래프는 기본적으로 화면을 꽉 채워서 보여주려는 특징이 있음
# 기준을 만들어줘야한다!

plt.plot(x, np.sin(x))      # 이 그래프로 그리면 x축과 y축의 기준이 다름..!
plt.xlim(-1,+11)            # x 축에 -1과 +11의 limit를 줌!
plt.ylim(-1.5,1.5)          # y 축에 -1.5와 +1.5의  limit을 줌
```

```python
# 그래프의 상하좌우 반전

# lim의 값을 최대, 최소로 지정한다.
plt.plot(x, np.sin(x))
plt.xlim(10,0)  # 좌우반전
plt.ylim(1.2,-1.2)  # 상하반전
```

```python
# 짧은 방법으로 표기

plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])  # x,y 의 좌표를 list type으로 만들어야한다.
```

```python
# Tight한 그래프 = 기본값

plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
# 실제 사이즈 그래프

plt.plot(x,np.sin(x))
plt.axis('equal')   # 1:1 사이즈의 그래프를 보여줌
```

```python
# label 붙이기

plt.plot(x, np.sin(x))

# 그래프의 제목을 붙여준다!
plt.title('A Sine Curve')

# x,y 축에 이름을 붙이기!
plt.xlabel('x')
plt.ylabel('sin(x)')

print('출력완료')
```

```python
# 그래프 label 붙이기!

plt.plot(x, np.sin(x), 'g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
# 하지만 label이 그래프에 보이지 않는다..

plt.legend()    # 각각의 label을 표시해줌!

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sin and Cos')
```

```python
# 산점도의 표현 (Scatter Plot)

plt.plot(x, np.sin(x),'o')  # marker 사용방법
```
